<h1>Estruturas de Seleção</h1>

<p>As estruturas de seleção são usadas para que sejam estabelecidos caminhos diferentes a partir de tomadas de decisão. Existem três tipos de estruturas de seleção: <strong>Simples, Composta e Encadeada</strong>. Cuja aplicação irá depender do contexto de utilização.</p>

<h2>Simples</h2>

<p> A estrutura de seleção simples faz uso da instrução if. Ela é utilizada quando queremos testar uma condição antes de executarmos uma ou mais instruções, por exemplo:
     
    idade = 20

    if idade >= 18:
        print("Você é maior de idade.")
</p>

<h2>Composta</h2>

<p> Temos também o else que é usado em conjunto com o if para fornecer uma alternativa caso a condição do if seja falsa. O bloco de código dentro do else é executado apenas se a condição do if for falsa, por exemplo:

    idade = 15
        
    if idade >= 18:
        print("Você é maior de idade.")
        else:
        print("Você é menor de idade.")
</p>

<h2>Encadeada</h2>

<p> Elif é uma abreviação de "else if" e permite verificar múltiplas condições. Se a condição de um if for falsa, o elif será utilizado. Se a condição do elif for verdadeira, o bloco de código associado a ele será executado, por exemplo:

    idade = 24
        
    if idade < 18:
         print("Você é menor de idade.")
    elif idade == 18:
         print("Você acabou de atingir a maioridade.")
    else:
        print("Você é maior de idade.")
</p>

<p> As estruturas de seleção permitem que você controle o fluxo do seu programa com diferentes condições, o que torna o código mais dinâmico e capaz de lidar com diversas situações. </p>

<h1>Estruturas de Repetição</h1>

<p>As estruturas de repetição são como um "faz tudo" do código. Elas permitem que você execute um bloco de código várias vezes, sem precisar ficar repetindo tudo.</p>

<p>Existem duas principais estruturas de repetição:

 <li>For: É utilizado quando você sabe quantas vezes precisa repetir o código. É ideal para percorrer listas, strings, tuplas e outras sequências.</li>

No exemplo abaixo o código dentro do loop for será executado 10 vezes, imprimindo os números de 1 a 10.

    for i in range(10):
        print(f"Número {i + 1}: {i}")


 <li>While: É utilizado quando você não sabe quantas vezes precisa repetir o código. O loop só vai parar quando uma condição for verdadeira.</li>

 No exemplo abaixo o código dentro do loop while será executado enquanto o valor da variável numero for menor ou igual a 20.  

    numero = 1
    while numero <= 20:
      print(f"Número {numero}")
    numero += 1   

 </p>

 <p>Outras dicas também, é utilizar os comandos: continue (que permite passar para a próxima iteração do loop, ignorando o restante do código dentro do loop) e break (que permite sair do loop completamente, antes que todas as iterações sejam concluídas).
